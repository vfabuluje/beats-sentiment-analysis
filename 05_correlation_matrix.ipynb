```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("earbuds_reviews_prepared_for_analysis.csv")

df["is_verified_numeric"] = df["is_verified"].astype(int)

numeric_df = df[["rating", "helpful_count", "review_length", "normalized_rating", "is_verified_numeric"]]

correlation_matrix = numeric_df.corr().round(2)
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()

sns.scatterplot(x="review_length", y="helpful_count", data=df, alpha=0.6)
plt.title("Review Length vs Helpful Count")
plt.xlabel("Review Length (words)")
plt.ylabel("Helpful Votes")
plt.show()

correlation_matrix.to_csv("correlation_matrix.csv", index=True)
```
