```python
# 02_misc_analysis

import numpy as np
import pandas as pd

df = pd.read_csv("earbuds_reviews_cleaned.csv")

ratings = df["rating"].values
mean_rating = np.nanmean(ratings)
ratings = np.where(np.isnan(ratings), mean_rating, ratings)
ratings = np.where((ratings >= 1) & (ratings <= 5), ratings, np.nan)
df["rating"] = ratings

df["content"] = df["content"].fillna("").astype(str)

review_lengths = np.array([len(review.split()) for review in df["content"]])
df["review_length"] = review_lengths

min_rating = np.nanmin(ratings)
max_rating = np.nanmax(ratings)
df["normalized_rating"] = (ratings - min_rating) / (max_rating - min_rating)

df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

df = df[df["product_name"].notna()]

df.to_csv("earbuds_reviews_prepared_for_analysis.csv", index=False)

df.info()
df.head()
```
